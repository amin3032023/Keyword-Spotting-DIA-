```python
from PIL import Image

def grayscale_conversion(image):

    return image.convert('L')

def binarization(image, threshold):

    grayscale_image = grayscale_conversion(image)

    pixels = grayscale_image.load()

    binary_image = Image.new('1', grayscale_image.size)

    binary_pixels = binary_image.load()

    # Iterate through each pixel and apply thresholding
    for i in range(grayscale_image.size[0]):
        for j in range(grayscale_image.size[1]):
            if pixels[i, j] >= threshold:
                binary_pixels[i, j] = 255
            else:
                binary_pixels[i, j] = 0

    return binary_image

input_image_path = "a1.jpg"
input_image = Image.open(input_image_path)

threshold_value = 127

# Perform binarization
binary_output = binarization(input_image, threshold_value)

# Save the binary image to a file
binary_output.save("binary_image.jpg")

print("Binary image saved as binary_image.jpg")
```

```python
from PIL import Image

image = Image.open('Doc1.jpg').convert('L')
width, height = image.size
total_intensity = sum(sum(image.getpixel((x, y)) for x in range(width)) for y in range(height))
global_threshold_value = total_intensity // (height * width)

binarized_image = Image.new('L', (width, height))
for y in range(height):
    for x in range(width):
        intensity = image.getpixel((x, y))
        if intensity >= global_threshold_value:
            binarized_image.putpixel((x, y), 255)
        else:
            binarized_image.putpixel((x, y), 0)


binarized_image.save('binarized_image1.jpg')
binarized_image.show()
```

```python

with open('a3.jpg', 'rb') as f:

    f.seek(2)
    image_data = f.read()
width = 256
height = 256

window_size = 3
binarized_image = [[0 for _ in range(width)] for _ in range(height)]

for y in range(height):
    for x in range(width):
        min_y = max(0, y - window_size // 2)
        max_y = min(height - 1, y + window_size // 2)
        min_x = max(0, x - window_size // 2)
        max_x = min(width - 1, x + window_size // 2)

        total_intensity = 0
        num_pixels = 0
        for yy in range(min_y, max_y + 1):
            for xx in range(min_x, max_x + 1):
                pixel_index = yy * width + xx
                total_intensity += image_data[pixel_index]
                num_pixels += 1
        local_threshold = total_intensity // num_pixels
        pixel_index = y * width + x
        if image_data[pixel_index] >= local_threshold:
            binarized_image[y][x] = 255
        else:
            binarized_image[y][x] = 0

for row in binarized_image:
    print(row)
```
